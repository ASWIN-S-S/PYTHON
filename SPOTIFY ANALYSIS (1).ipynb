### 7 Which are the unique genres in the dataset?

```python
spoty["playlist_genre"].unique()
```

### 8 Calculate the average duration of songs in minutes (convert 'duration_ms' to minutes).

```python
minutes = spoty["duration_ms"].apply(lambda x:x/60000)
average = minutes.mean()
print("The average duration of songs in minutes is",average)
```

### 9 Count the number of songs released each year and plot it using a line plot.

```python
num_songs = spoty["Year"].value_counts().sort_index()
num_songs
```

```python
sns.lineplot(data=num_songs)
```

### 10 Calculate the total duration of songs for each genre and plot it using a bar chart.

```python

total_duration_by_genre = spoty.groupby('playlist_genre')['duration_ms'].sum().sort_values()

# Convert duration from milliseconds to hours for better readability
total_duration_by_genre_hours = total_duration_by_genre / (1000 * 60 * 60)
total_duration_by_genre_hours
```

```python



plt.figure(figsize=(10, 6))
sns.barplot(x=total_duration_by_genre_hours.index, y=total_duration_by_genre_hours.values, palette="viridis")
plt.title('Total Duration of Songs by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Duration (Hours)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 11 Create a scatter plot of danceability versus energy, colored by genre.

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=spoty, x='danceability', y='energy', hue='playlist_genre', palette='tab10', alpha=0.7)
plt.title('Danceability vs Energy by Genre')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.legend(title='Genre',  loc='upper left')
plt.tight_layout()
plt.show()
```

### 12 Create a column 'tempo_category' that categorizes songs as 'Slow', 'Moderate', or 'Fast' based on their tempo (Slow: < 90 BPM, Moderate: 90-140 BPM, Fast: > 140 BPM).


```python

def categorize_tempo(tempo):
    if tempo < 90:
        return 'Slow'
    elif 90 <= tempo <= 140:
        return 'Moderate'
    else:
        return 'Fast'


spoty['tempo_category'] = spoty['tempo'].apply(categorize_tempo)
spoty[['track_name', 'tempo', 'tempo_category']].head()
```

### 13 Find the artist with the highest average popularity across all their songs.

```python
average_popularity_by_artist = spoty.groupby('track_artist')['track_popularity'].mean()
most_popular_artist = average_popularity_by_artist.idxmax()
highest_average_popularity = average_popularity_by_artist.max()

print(f"The artist with the highest average popularity is {most_popular_artist} with an average popularity of {highest_average_popularity}.")
```

### 14  Identify duplicate songs based on 'name' and 'artist'. How many duplicate songs are there?

```python
duplicates = spoty[spoty.duplicated(subset=['track_name', 'track_artist'], keep=False)]
num_duplicates = duplicates.shape[0] - duplicates['track_id'].nunique()
print(f'There are {num_duplicates} duplicate songs.')
```

```python
duplicate_songs = spoty[spoty.duplicated(subset=['track_name', 'track_artist'], keep=False)]

num_duplicates = duplicate_songs.shape[0]- duplicate_songs["track_id"].nunique()
print(f"There are {num_duplicates} duplicate songs based on 'track_name' and 'track_artist'.")
```

### 18  Create a FacetGrid of histograms showing the distribution of 'acousticness' for each 'tempo_category'.

```python
g = sns.FacetGrid(spoty, col='tempo_category', col_wrap=3, height=4, aspect=1.5)
g.map(sns.histplot, 'acousticness', kde=False, bins=30, color='blue')
g.set_titles('{col_name} Tempo')
g.set_axis_labels('Acousticness', 'Frequency')
plt.suptitle('Distribution of Acousticness by Tempo Category', y=1.05)
plt.tight_layout()
plt.show()
```

### 19  Based on the analysis, what are the key factors that influence the popularity of a song?

```python
correlation_matrix = spoty.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix[['track_popularity']].sort_values(by='track_popularity', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Track Popularity')
plt.show()
```
